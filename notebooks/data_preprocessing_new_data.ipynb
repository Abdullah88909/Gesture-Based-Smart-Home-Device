```python
import os, shutil

# correct dataset root
src_root = "/content/hagrid_raw/hagrid-classification-512p-no-gesture-150k"
dst_root = "/content/hagrid_clean"

keep = ["palm", "fist", "like", "dislike"]

os.makedirs(dst_root, exist_ok=True)

for cls in keep:
    src = os.path.join(src_root, cls)
    dst = os.path.join(dst_root, cls)
    print(f"Copying {src} -> {dst}")
    shutil.copytree(src, dst, dirs_exist_ok=True)

print("\nDONE  clean 4-class dataset created at /content/hagrid_clean")
```
